```python

# Dados simulados
dados_simulados_auditoria = {
    'Data e Hora': ['2024-02-03 10:15:00', '2024-02-03 11:30:45', '2024-02-03 14:20:30'],
    'Número de Transação': [123456789, 987654321, 456789012],
    'Tipo de Transação': ['Transferência', 'Pagamento', 'Depósito'],
    'Conta Origem': ['Conta123', 'Conta789', 'Conta789'],
    'Conta Destino': ['Conta456', 'FornecedorXYZ', 'AgênciaXYZ'],
    'Valor': [500.00, 150.00, 1000.00],
    'Status': ['Concluída', 'Pendente', 'Concluído'],
    'Detalhes': ['ID da Sessão: ABC123, IP: 192.168.1.10',
                 'ID da Sessão: DEF456, IP: 192.168.1.15',
                 'ID da Sessão: GHI789, IP: 192.168.1.20']
}

# Criando o DataFrame
df_logs_auditoria = pd.DataFrame(dados_simulados_auditoria)

# Exibindo o DataFrame
df_logs_auditoria.head()
```

```python
# Criando o DataFrame
df_logs_auditoria = pd.DataFrame(dados_simulados_auditoria)

# Calculando estatísticas descritivas
media = df_logs_auditoria['Valor'].mean()
mediana = df_logs_auditoria['Valor'].median()
moda = df_logs_auditoria['Valor'].mode().values[0]
media_geometrica = (df_logs_auditoria['Valor'] + 1).prod() ** (1 / len(df_logs_auditoria['Valor'])) - 1

# Exibindo estatísticas
print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Moda: {moda}')
print(f'Média Geométrica: {media_geometrica}')

# Criando um box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df_logs_auditoria['Valor'], vert=False)
plt.title('Box Plot - Valores')
plt.xlabel('Valor')
plt.show()
```

```python


# Vamos criar mais dados simulados para ter diferentes valores para cada tipo de transação
dados_simulados_auditoria['Valor'] = [500.00, 150.00, 1000.00, 300.00, 800.00, 120.00, 1500.00, 900.00, 600.00]

# Realizando o teste t entre diferentes tipos de transação
teste_t, valor_p = ttest_ind(df_logs_auditoria[df_logs_auditoria['Tipo de Transação'] == 'Transferência']['Valor'],
                              df_logs_auditoria[df_logs_auditoria['Tipo de Transação'] == 'Pagamento']['Valor'])

# Exibindo os resultados
print(f'Resultado do Teste t: {teste_t}')
print(f'Valor p: {valor_p}')

# Box plot para visualização
plt.figure(figsize=(8, 6))
sns.boxplot(x='Tipo de Transação', y='Valor', data=df_logs_auditoria)
plt.title('Box Plot - Valores por Tipo de Transação')
plt.show()
```
